## Plot of phytoplankton cell growth data from Flynn94:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
phytoplankton_growth_data = pd.read_csv('~/Documents/HonsUCT_IntroOceMod/data/flynn94.csv')
```

```python
plt.plot(phytoplankton_growth_data['time'], phytoplankton_growth_data['cells'])
plt.xlabel('Time (days)')
plt.ylabel('Phytoplankton Cells ($10^6 {ml}^{-1}$)')
plt.title('Phytoplankton Cell Growth Over Time')
plt.show()
```

**Question 1:** Is the growth exponential?
Yes the growth appears to be exponential.

**Question 2:** Can you separate the growth curve into some phases?
The curve can be split into 3 distinct phases: the initial phase where the growth is slow and the curve is quite flat, then the phase where the growth is rapid (exponential) and the curve is a steep incline, and then the phase where the growth basically stops and plateaus at around $2.00$ x$10^6 ml^{-1}$ cells, indicating growth has become limited by resources.

**Question 4:** Is the ammonium consumption approximated by an exponential curve?

```python
plt.plot(phytoplankton_growth_data['time'], phytoplankton_growth_data['ammonium'])
plt.xlabel('Time (days)')
plt.ylabel(r'Ammonium ($\mu$g N ml$^{-1}$)')
plt.title('Ammonium Concentration Over Time')
plt.show()
```

**Question 5:** How would you estimate the parameters of the growth curve?

Using Malthusian growth model:
The increase of phytoplankton cells over time can be expressed as $\Delta N = \mu N \Delta t$ where the growth rate $\mu$ is in units of 1/time.

Putting this into differential form we obtain the dynamics of the exponential growth model:

$\frac{dN}{dt} = \mu N$

which can be solved analytically by seperating the variables, integrating both sides and applying an initial condition @ $t=0, N(t=0)=N_0$:

$\int \frac{1}{N}dN = \mu \int dt \Rightarrow N = N_0e^{\mu t}$
