# The Lotka-Volterra model
This model is a python translation of the LV_pz model proposed in Glover et al. (2011), Sec. 9.3

You will need to make sure the required modules below are installed in your python environment

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

This is the function containing the right hand side of the ODE system

```python
def LV_pz(x,t,p):
    X1,X2 = x
    p1,p2,p3,p4 = p
    dX1dt = X1 * (p1 - p2*X2)
    dX2dt = X2 * (p3*p2*X1 - p4)
    return np.array([dX1dt,dX2dt])
```

## Parameters and initial conditions

```python
# The parameters
p = np.array([0.1,0.4,0.2,0.05])
X1_0 = 0.8
X2_0 = 0.171
x0 = np.array([X1_0,X2_0])
```

## Time axis

```python
# time units are days
Tstart = 0.
Tend = 400.
step = 0.1
t = np.arange(Tstart,Tend,step)
```

## Integration using the scipy ODE solver

```python
# note that the extra argument containing the parameter "args" must be a tuple
y = odeint(LV_pz, x0, t, args=(p,))
```

## Plot Fig. 9.3a

```python
def plot_plankton_vs_time(y, t):
    plt.plot(t,y[:,0],'k-',label='Phytoplankton')
    plt.plot(t,y[:,1],'k--',label='Zooplankton')
    plt.xlabel('Time (days)')
    plt.ylabel('Plankton (mmol N/m3)')
    plt.legend()
    plt.show()
```

```python
plot_plankton_vs_time(y, t)
```


## 1. Write the code to plot Fig. 9.3b, the trajectory in the phase space

```python
def plot_phase_space_trajectory(y):
    plt.plot(y[:,0],y[:,1], color="black")
    plt.xlabel('Phytoplankton ($mmol N m^{-3}$)')
    plt.ylabel('Zooplankton ($mmol N m^{-3}$)')
    plt.title("Phytoplankton & Zooplankton trajectory in the phase space")
    plt.show()
```

```python
plot_phase_space_trajectory(y)
```

## 2. Set the parameter $p_3 = 1$ and compare how different the trajectory is in the phase space

```python
p = np.array([0.1,0.4,1,0.05])
y = odeint(LV_pz, x0, t, args=(p,))

plot_phase_space_trajectory(y)
```

The trajectory seems to become larger and less circular when $p_3=1$. The axis both start from 0 in comparison to the original plot, and for zooplankton, a higher concentration seems to be reached.

## 3. Write the code to compute the solution of the L-V model using the initial conditions given in the caption of Fig. 9.4 & plot the results

Parameters of the model:
- State variables, $x_1$ and $x_2$ represent the population densities of prey and predator, respectively.
- The constant of proportionality (a growth rate) is $p_1$.
- Constant of proportionality (related to predation success) is $p_2$.
- Assimilation efficiency is $p_3$ and will be between 0 and 1.
- $p_4$ represents the mortality rate of the predator.
- The parameters $p_i$ control the behavior of the model.
- This model doesn’t really conserve mass, there is no limit to growth in the prey population and the efficiency of conversion of prey to predator mass ( $p_3$ ) will likely be less than 100%.


The coupled ODEs describing this model are based on the concept of consveration of mass among state variables- flux into one of the boxes results in an increase in population density ($x_i$ variable), and any flux out of the box leads to a decrease.

$\frac{dx_1}{dt} = p_1x_1 - p_2x_1x_2$

$\frac{dx_2}{dt} = p_2p_3x_1x_2 - p_4x_2$

**From Figure 9.4 caption:** 
- Initial condition vector is now [0.62, 0.25].
- Time integration is 1100 days (roughly three years).

```python
#initial conditions from fig 9.4
X1_0 = 0.62
X2_0 = 0.25
x0 = np.array([X1_0,X2_0])

#time = 1100 days
Tstart = 0
Tend = 1100
step = 0.1
t = np.arange(Tstart,Tend,step)

#p remains the same as original
p = np.array([0.1,0.4,0.2,0.05])

y = odeint(LV_pz, x0, t, args=(p,))
plot_plankton_vs_time(y, t)
plot_phase_space_trajectory(y)
```

## 4. Why are they different? Think about a few hypotheses and then see the explanation in the document Glover_etal_2011_ERRATA.pdf

The results plotted above appear to be quite different from those in fig 9.4 in the glover et al paper, with very shallow almost flat oscillation happening in the first plot and a symmetrical oval formed in the second plot. The plot above shows a closed phase loop and oscillation. In the paper, their results appeared different due to an unfavorable interaction between truncation and roundoff error accumulating in the solution.
