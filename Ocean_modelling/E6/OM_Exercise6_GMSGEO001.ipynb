```python
import pylab as pl
import matplotlib.pyplot as plt
import numpy as np
```

```python

#Euler implicit backward solution for decay ODE

def decay_ODE_euler_backward_sol(t0, tn, dt, kappa, C0):
    def F(k, C): # define the decay function for RHS
        return -k*C
        
    # compute the total number of steps:
    Ntot = (pl.floor(( tn-t0 )/dt) + 1).astype(int)

    Ntot
    
    #initialization
    C = pl.zeros(Ntot)
    t = C.copy( ) 
    t[0] = t0
    C[0] = C0
    
    #start the loop
    for n in range(Ntot-1):
        C[n+1] = C[n]/(1 + (kappa * dt)) 
        t[n+1] = t[n] + dt
  
    return C, t  
```

```python
#constants
t0 = 0 # time is in days
tn = 20
kappa = 0.05 # day1
C0 = 0.5 # mg chl/m3

#results for dt = 10
C_delta10, t_delta10 = decay_ODE_euler_backward_sol(t0, tn, 10, kappa, C0)

plt.plot(t_delta10, C_delta10, 'o-', label=r'$\Delta t$ = 10 days')
plt.xlabel('Time [days]')
plt.ylabel(r'Chlorophyll [$mg/m^3$]')
plt.legend()


#results for dt = 5
C_delta5, t_delta5 = decay_ODE_euler_backward_sol(t0, tn, 5, kappa, C0)

plt.plot(t_delta5, C_delta5, 'o-', label=r'$\Delta t$ = 5 days')
plt.xlabel('Time [days]')
plt.ylabel(r'Chlorophyll [$mg/m^3$]')
plt.legend()

#results for dt = 1
C_delta1, t_delta1 = decay_ODE_euler_backward_sol(t0, tn, 1, kappa, C0)

plt.plot(t_delta1, C_delta1, 'o-', label=r'$\Delta t$ = 1 day')
plt.xlabel('Time [days]')
plt.ylabel(r'Chlorophyll [$mg/m^3$]')
plt.legend()


#show results 
plt.show()
```

As the time step gets smaller, the approximation of decay becomes smoother.
