**Parameters**: 
- Time axis in days, from 0 to 20 days
- $k' = 0.1 d^{−1}$
- $P_0 = 0.1  mmol N m^{−3}$
- $N_0 = 5  mmol N m^{−3}$

**Model**:  

$k = k'N$

$\frac{dP}{dt} = k'NP$

$\frac{dN}{dt} = -k'NP$ 

**Analytical Solution**:  

$N(t) = P_0 + N_0 - P(t)$ 

$P(t) = P_0\frac{P_0 + N_0}{P_0 + N_0e^{−k′P_0+N_0t}}$ 


```python

import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#initial parameters
P0 = 0.1
N0 = 5
k = 0.1
t = np.arange(0, 21, 1) #1D timestep array up to 20 days  
```

```python
def plot_growth_vs_consumption(N0, P0, k, t):

    P = np.zeros(len(t))   # Empty array to store Phytoplankton concentration values
    P[0] = P0
    N = np.zeros(len(t))   # Empty array to store Nutrient concentration values
    N[0] = N0

    #Implement analytical solution to get P(t) and N(t) values for t=1,...,20 & store in seperate arrays
    for j in range(1, len(t)):
        time = t[j]
        P_t = P0*((P0 + N0)/(P0 + (N0*math.exp(-1*k*(P0+N0)*(time)))))
        N_t = P0 + N0 - P_t
        P[j] = P_t
        N[j] = N_t
        

    #Plotting the trajectory for each concentration over time:
    growth_plot = plt.plot(t, P, color="green", label = 'P')
    consumption_plot = plt.plot(t, N, color="red", label = 'N')
    plt.xticks(np.arange(0, max(t)+1, 2)) 
    plt.xlabel ('Time (days)')
    plt.ylabel (r'$mmol  N  m^{−3}$')
    plt.title("Phytoplankton Growth vs Nutrient Consumption")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
plot_growth_vs_consumption(N0, P0, k, t)
```

## Question 1: Why are the curves identical in shape?

The two curves are identical in shape because of the mathematics; they are directly proportional to one another and are both exponential curves. They change with respect to the same constant but one is in decay and one is growing. Phytoplankton will continue to consume nutrients and grow up until the point where the nutrients run out and then growth stops and levels off at about $5 mmol N m^{-3}$ and the nutrient concentration levels off at 0 because it has been fully used. The use of nutrients is rapid initially and the growth is also rapid initially before this levelling off which is typically of exponential curves.
